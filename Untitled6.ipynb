```python
import numpy as np


# Код Хэ́мминга — самоконтролирующийся и самокорректирующийся код. Построен применительно к двоичной системе счисления. 
# Позволяет исправлять одиночную ошибку (ошибка в одном бите слова) и находить двойную. Назван в честь американского математика 
# Ричарда Хэмминга, предложившего код.

option = int(input(
    'Нажмите 1 для генерации кода Хэмминга  \nНажмите 2 для поиска ошибки в коде Хэмминга\n\t Введите свой выбор:-> '))


def generate(d):
    data = list(d)
    print("data->", *data)
    data.reverse()
    c, ch, j, r, h = 0, 0, 0, 0, []
    while (len(data) + r + 1) > (pow(2, r)):
        r = r + 1

    for i in range(0, (r + len(data))):
        p = (2 ** c)
        # print(p)

        if p == (i + 1):
            h.append(0)
            c += 1
            # print(c)

        else:
            h.append(int(data[j]))
            j += 1
            # print(h)

    for parity in range(0, (len(h))):
        ph = (2 ** ch)
        if ph == (parity + 1):
            startIndex = ph - 1
            i = startIndex
            toXor = []

            while (i < len(h)):
                block = h[i:i + ph]
                toXor.extend(block)
                i += 2 * ph
                # print(toXor)

            for z in range(1, len(toXor)):
                h[startIndex] = h[startIndex] ^ toXor[z]
            ch += 1

    h.reverse()
    print('Сгенерированный код Хэмминга будет выглядеть следующим образом:- ', end="")
    print(int(''.join(map(str, h))))


def detection(d):
    data = list(str(d))
    print('data->', data)
    data.reverse()
    c, ch, j, r, error, h, parity_list, h_copy = 0, 0, 0, 0, 0, [], [], []

    for k in range(0, len(data)):
        p = (2 ** c)
        h.append(int(data[k]))
        h_copy.append(data[k])
        # print(h)
        if p == (k + 1):
            c = c + 1

    for parity in range(0, (len(h))):
        ph = (2 ** ch)
        if ph == (parity + 1):

            startIndex = ph - 1
            i = startIndex
            toXor = []

            while i < len(h):
                block = h[i:i + ph]
                toXor.extend(block)
                i += 2 * ph

            for z in range(1, len(toXor)):
                h[startIndex] = h[startIndex] ^ toXor[z]
            parity_list.append(h[parity])
            ch += 1
    parity_list.reverse()
    error = sum(int(parity_list) * (2 ** i) for i, parity_list in enumerate(parity_list[::-1]))

    if error == 0:
        print('В полученном коде Хэмминга нет ошибки')

    elif error >= len(h_copy):
        print('Ошибка не может быть обнаружена')

    else:
        print('Ошибка заключается в', error, 'бит')

        if h_copy[error - 1] == '0':
            h_copy[error - 1] = '1'

        elif h_copy[error - 1] == '1':
            h_copy[error - 1] = '0'
            print('Код Хемминга после исправления:-> ')
        h_copy.reverse()
        print(int(''.join(map(str, h_copy))))


def menu():
    if option == 1:  # ГЕНЕРИРОВАТЬ КОД ХЭММИНГА
        choice = int(
            input('Нажмите 1 для генерации рандомной матрицы NxN\n Нажмите 2 для ввода вашей матрицы\n Ваш выбор:-> '))
        if choice == 1:
            a1 = int(input('Введите колличество стобцов: -> '))
            print('*' * 10)
            a2 = int(input('Введите колличество векторов: -> '))
            d = np.random.randint(0, 2, (a1, a2))
            print("d->", d)
            print("*" * 30)
            cd = np.squeeze(np.array(d))
            print('cd->', cd)
            for m in cd:
                generate(m)


        elif choice == 2:
            print('Поддерживает высоту только 4 столца\n', '*' * 10)
            a1 = int(input('Введите 1-й вектор: -> '))
            a2 = int(input('Введите 2-й вектор: -> '))
            a3 = int(input('Введите 3-й вектор: -> '))
            a4 = int(input('Введите 4-й вектор: -> '))
            d = np.vstack([a1, a2, a3, a4])
            print('d->', d)
            # cd = np.squeeze(np.asarray(d))
            # print("cd->", cd)
            for m in d:
                generate(m)

        else:
            print('Выберите из предложенных вариантов:')


    elif option == 2:  # ОБНАРУЖЕНИЕ ОШИБКИ В ПОЛУЧЕННОМ КОДЕ ХЭММИНГА
        print('Введите полученный код Хэмминга по векторам')
        print('Поддерживает размерность только 4 столбца')
        d = []
        a1 = int(input('Ввдеите 1-й код Хемминга: -> '))
        d.append(a1)
        a2 = int(input('Ввдеите 2-й код Хемминга: -> '))
        d.append(a2)
        a3 = int(input('Ввдеите 3-й код Хемминга: -> '))
        d.append(a3)
        a4 = int(input('Ввдеите 4-й код Хемминга: -> '))
        # d.append(a4)
        # d = np.vstack([a1, a2, a3, a4])
        print("d->", d)
        # cd = np.squeeze(np.asarray(d))
        # print('cd->', cd)
        for m in d:
            print(m)
            detection(m)


    else:
        print('Введенный параметр не существует')


if __name__ == '__main__':
    menu()
```
